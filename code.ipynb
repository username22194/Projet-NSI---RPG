```python
def VerifInput(dictionnaire, variable):
    for element in dictionnaire.items:
        while variable < 1 or variable > len(dictionnaire):
            print("Veuillez entrer un nombre valide.")
            variable = int(input())
    return variable
```

```python
Personnages = {'1':'Salie', '2':'Magnus'}

statsSalie = {'HP':100, 'DEF':5, 'DMG':12, 'critluck':0.25, 'luck':0.75, 'EXP':0, 'LVL':1}
ActionsSalie = {'1':'Attaque basique', '2':'Lancer un projectile', '3':'Utiliser un consommable', '4':'Fuire'}
Projectiles = {'1':'Tonneau explosif', '2':'Boule de feu', '3':"Potion jetable d'acide", '4':'Parfum de poison'}

statsMagnus = {'HP':150, 'DEF':10, 'DMG':10, 'critluck':0.30, 'luck':0.6, 'EXP':0, 'LVL':1}
ActionsMagnus = {'1':'Attaque basique', '2':'Parer (3 tours)', '3':"Lancer l'arme en main (2x Dégâts mais vous faites ensuite moins de Dégâts)", '4':'Utiliser un consommable', '5':'Fuire'}

PotionDeSoin = {'1':'Petite potion de soin', '2':'Potion de soin moyenne', '3':'Grande potion de soin'}
Consommables = {'1':'Potion de soin(petite/moyenne/grande)', '2':'Potion de régénération', '3':'Potion de défense', '4':'Potion de rage', '5':'Piment'}

CombatStats = {}
```

```python
def ChoisirHéros(): #Choisir le héros (entre Salie et Magnus)
    print("Vous avez deux choix de personnage :")
    for element in Personnages.items():
        print(f"{Personnages.keys}: {Personnages.values}")

    global ChoixHéros
    ChoixHéros = int(input('Choisissez votre personnage '))
    VerifInput(Personnages, ChoixHéros)
    
    global stats
    global ActionsDeCombat
    global statstoAppend
    global CombatStats

    if ChoixHéros == 1: 
        stats = statsSalie
        ActionsDeCombat = ActionsSalie
        print("Vous avez choisi Salie.")

    if ChoixHéros == 2: 
        stats = statsMagnus
        ActionsDeCombat = ActionsMagnus
        print("Vous avez choisi Marcus.")
    
    statstoAppend = ['HP', 'DEF', 'DMG']
    CombatStats = {key: stats[key] for key in statstoAppend}
                            
    print("Voici vos statistiques (Elles augmentent avec le niveau) : ")
    for key, value in stats.items():
        print(f"{key}: {value}")

    return ChoixHéros
```

```python
import random as rd
def Critique(critluck, dmg): # L'attaque est-elle un coup critique / les dégâts ?
    global crit
    global DamageDealt
    crit = False
    critdmg = rd.random() + 1
    if rd.random() < stats[critluck]:
        crit = True
        DamageDealt = int(dmg*critdmg)
    else:
        DamageDealt = int(dmg)
```
